# Q.1 Import the libraries




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

# Q.2 Get the Dataset

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

# Q3. Plot the distribution of total_bill in the given dataset.

```python
sns.distplot(tips['total_bill']);
```

The distribution is right skewed. The sum of the curve is equal to one. Maximum number of distribution is between 13.33 and 16.67. 

# Q.4 - Plot the distribution plot for tips.

```python
sns.distplot(tips['tip']);
```

Maximum frequency of tips is between 2 to 2.5. The distribution is again right skewed.

# Q5. Make a joint plot between tip and total bill.

```python
sns.jointplot(x='tip', y='total_bill', data=tips);
```

The bivariate relationship between tip and total_bill is mostly linear. 

# Q6. Give additional attribute kind = "hex" in the above plot.

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex');
```

# Q7. Make pairplot between total bill and tip.

```python
sns.pairplot(tips, vars=['tip', 'total_bill']);
```

# Q.8 -  Make striplot between sex and tip.

```python
sns.stripplot(x='sex', y='tip', data=tips, jitter=True);
```

A stripplot is build between a categorical and a continuous data. We can observe from the graph that the tip is similar for both male and female with some outliers in case of male.

# Q.9 - What other plots you can make for tip and sex variable.Plot them.

```python
sns.boxplot(x='sex', y='tip', data=tips, hue='smoker');
```

The interquartile range for male is greater than female. There are some outliers in case of male non-smoker. The medians in case of smoker is higher than non-smokers for both sexes.

# Q.10 Make bar plot with day and total bill.

```python
sns.barplot(x='day', y='total_bill', data=tips);
print('')
print(tips.shape)
```

Total bill is highest for Sun and lowest for Fri. The error bars potray 95% of accuracy of the mean() of the distribution and is termed as the confidence interval.

# Q.11 - Make a plot to count total males and females for each day.

```python
sns.countplot(x='day', hue='sex', data=tips);
```

The maximum number of males are on sun and lowest on Fri. The maximum number of females are on Thur and lowest on Fri.
The maximum number of males is close to 60 and least is 10. The maximum number of females is 32 and least is 9.

# Q.12 -  Use factorplot to plot multiple categorical variables like day, smoker, peoplecount.

```python
sns.catplot(x='day',  hue='sex', col='smoker', data=tips, kind='count');
```

# Q.13 - Use lmplot() to plot the relation between total bill and tip.


```python
sns.lmplot(x='tip', y='total_bill', data=tips, hue='sex');
```
